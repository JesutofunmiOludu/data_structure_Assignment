#**DATA STRUCTURE IN PYTHONS**

```python
# List: Ordered, mutable, allows duplicate elements.
my_list = [1, 2, 3, 'apple', 'banana', 3]
print(f"List: {my_list}")

# Tuple: Ordered, immutable, allows duplicate elements.
my_tuple = (10, 20, 'cherry', 'date', 10)
print(f"Tuple: {my_tuple}")

# Dictionary: Unordered, mutable, stores key-value pairs.
my_dict = {
    'name': 'Alice',
    'age': 30,
    'city': 'New York'
}
print(f"Dictionary: {my_dict}")

# Set: Unordered, mutable, stores unique elements.
my_set = {1, 2, 3, 'apple', 'banana', 2, 1}
print(f"Set: {my_set}")
```
